```python
import pandas as pd
import numpy as np

df3['hsn_code'] = pd.to_numeric(df3['hsn_code'], errors='coerce')  # Convert non-numeric values to NaN
df3['hsn_code'] = df3['hsn_code'].replace([np.inf, -np.inf], np.nan).fillna(0).astype(int)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd


# Initialize the WebDriver (assuming you are using Chrome)
driver = webdriver.Chrome()

try:
    # Open the website
    driver.get('https://vakilsearch.com/hsn-code')

    # Iterate through each HSN code in the DataFrame
    for hsn_code_to_search in df3['hsn_code']:
        # Locate the search input element and enter the HSN code
        search_input = driver.find_element(By.CSS_SELECTOR, 'input[placeholder="Enter HS code or product description"]')
        search_input.clear()
        search_input.send_keys(hsn_code_to_search)

        # Press Enter to perform the search
        search_input.send_keys(Keys.RETURN)

        # Wait for the new page to load (adjust the time as needed)
        time.sleep(2)

        # Extract the product description from the new page
        try:
            description_element = driver.find_element(By.XPATH, f'//td[text()="{hsn_code_to_search}"]/following-sibling::td[1]')
            hsn_code_description = description_element.text
        except:
            product_description = "Not Found"

        # Append the result to the DataFrame
        df3.loc[df3['hsn_code'] == hsn_code_to_search, 'hsn_code_description'] = hsn_code_description

        # Print the result
        print(f" Description for HSN Code {hsn_code_to_search}: {hsn_code_description}")

finally:
    # Close the WebDriver
    driver.quit()

# Display the updated DataFrame
print(df3)
```
